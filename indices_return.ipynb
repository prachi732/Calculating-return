```
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```
tickers=['^GSPC','^IXIC','^GDAXI','^DJI']
data=pd.DataFrame()
for t in tickers:
  data[t]=wb.DataReader(t,data_source='yahoo',start='1997-1-1')['Adj Close']
```

```
returns=(data/data.shift(1))-1
returns
```
